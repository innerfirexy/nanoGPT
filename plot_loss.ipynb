```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Read loss data
def read_loss_from_log(log_file):
    losses = []
    with open(log_file, 'r') as f:
        for line in f:
            if 'val loss' in line:
                losses.append(float(line.split('val loss')[1].strip()))
    return losses

shkspr_char_loss = read_loss_from_log('./out-shakespeare-char/log.txt')
print(shkspr_char_loss)

shkspr_char_loss_cepred = read_loss_from_log('./out-shakespeare-char-cepred/log_0.5.txt')
print(shkspr_char_loss_cepred)

shkspr_char_loss_cepred2 = read_loss_from_log('./out-shakespeare-char-cepred/log_0.9.txt')
print(shkspr_char_loss_cepred2)

shkspr_char_loss_cepred3 = read_loss_from_log('./out-shakespeare-char-cepred/log_0.1.txt')
print(shkspr_char_loss_cepred3)
```

```python
# Plot loss curves

plt.plot(shkspr_char_loss[2:], label='Char')
plt.plot(shkspr_char_loss_cepred[2:], label='Char + CEPred 0.5')
plt.plot(shkspr_char_loss_cepred2[2:], label='Char + CEPred 0.9')
plt.plot(shkspr_char_loss_cepred3[2:], label='Char + CEPred 0.1')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
